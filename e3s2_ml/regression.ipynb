```python
class linear_regression_mse:
    def __init__(self,x,y,alpha=0.001,epochs = 10):
        self.m = 0.0
        self.c = 0.0
        self.regress(x,y,alpha,epochs)

    def regress(self,x,y,alpha,epochs):
        n = len(x)
        print("epoch\tm\t\tc\t\tloss\t\tdm\t\tdc\t\tnew_m\tnew_c")
        for i in range(epochs):
            cost_square_sum = 0.0 
            dm_sum = 0.0
            dc_sum = 0.0
            for j in range(n):
                yi = (x[j]*self.m)+self.c
                #ya = y[j]#y actual
                cost = y[j]-yi
                cost_square_sum = cost_square_sum + (cost*cost)
                dm_sum = dm_sum + cost*(-x[j]) 
                dc_sum = dc_sum + (-cost)
            loss = cost_square_sum/(2*n)
            dm = dm_sum/n
            dc = dc_sum/n
            old_m = self.m
            old_c = self.c
            self.m = self.m - (alpha*dm)
            self.c = self.c - (alpha*dc)
            print(f"{i+1}\t{old_m:.3f}\t\t{old_c:.3f}\t\t{loss:.3f}\t\t{dm:.3f}\t\t{dc:.3f}\t\t{self.m:.3f}\t{self.c:.3f}")
        return (self.m,self.c)
    
    def predict(self,x):
        y = []
        for j in range(len(x)):
            yi = (x[j]*self.m)+self.c
            y.append(yi)
        return y

```

```python
x = [20,30,40,25]
y = [30,40,50,30]

lrm = linear_regression_mse(x,y)
y_pred = lrm.predict(x)
print()
print("y_actu:",y)
print("y_pred:",y_pred)
```
