```python
from sklearn.datasets import make_blobs
X , y = make_blobs(n_samples=1000,n_features=2,centers=4,random_state=7)

import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c = y)
```

```python
from sklearn.preprocessing import StandardScaler
st_scaler = StandardScaler()

from sklearn.model_selection import train_test_split
X_train,X_test ,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
X_train_s = st_scaler.fit_transform(X_train)
X_test_s = st_scaler.transform(X_test)
from sklearn.cluster import KMeans

wcss = {}
for i in range(1,11):
    km = KMeans(n_clusters=i,init="k-means++")
    km.fit(X_train_s)
    wcss[i] = km.inertia_
print(wcss)
plt.plot(wcss.keys(),wcss.values())
plt.show()
# print(wcss.keys())
```

```python
km=KMeans(n_clusters=4,init="k-means++")
km.fit(X_train_s)
y_test_pred = km.predict(X_test_s)
y_test_pred
```
